```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('all_regions_data.csv')
```

```python
data
```

```python
shifted_arrivals= []
for place in data['place'].unique():
    shifted_arrivals.extend(data[data['place']==place]['monthly_arrivals'].shift(-1).values)
```

```python
data['monthly_arrivals_s'] = shifted_arrivals
```

```python
data.dropna(subset=['monthly_arrivals_s'],inplace=True)
```

```python
data['monthly_arrivals_s'] = data['monthly_arrivals_s'].astype(int)
```

```python
data
```

```python
# data = data.set_index("place")
# data = data.drop("banadir/mogadishu", axis = 0)
```

```python
data = data.iloc[44:]
```

```python
for col in data.columns:
    data[col] = data[col].replace("No Data",np.nan)
    data[col] = data[col].replace("",np.nan)
    data[col] = data[col].replace(" ",np.nan)
    data[col] = data[col].replace("-",np.nan)
```

```python
data = data.reset_index()
```

```python
data = data.drop(["index"], axis = 1)
```

```python
data.isna().sum()
```

```python
# data["climate/rainfall"] = data["climate/rainfall"].astype(float)
# data['climate/rainfall'] = data.groupby('place')['climate/rainfall'].apply(lambda x: x.fillna(x.mean()))
```

```python
data.columns
```

```python
for col in data.columns[2:21]:
    data[col] = data[col].astype(float)
    data[col] = data.groupby('place')[col].apply(lambda x: x.fillna(x.mean()))
```

```python
data.isna().sum()
```

```python
data.groupby("place").count().rsub(data.groupby("place").size(), axis = 0)
```

```python
sns.heatmap(data.isnull())
```

```python
final_cols = []
for col in data.columns:
    per = np.round(pd.isnull(data[col]).sum()/data.shape[0]*100,2)
    print("percentage of null points in column-%s are: %0.2f"%(col, per))
    if per < 30:
        final_cols.append(col)
```

```python
final_data = data[final_cols].copy()
```

##### 'woqooyi_galbeed/hargeysa'

```python
d1.head(2)
```

```python
d1[' Number of Individuals '] = d1[' Number of Individuals '].apply(lambda x: int(x.replace(",","")))
```

```python
d1['Month End'] = pd.to_datetime(d1['Month End'])
```

```python
d1[d1['Current (Arrival) District'].str.lower() == 'hargeysa'].groupby('Month End').sum().sort_values(by='Month End')
```

```python
hargesa_data = d2[d2['place']=='woqooyi_galbeed/hargeysa'].iloc[:-1].copy()
```

```python
hargesa_data.shape
```

```python
hargesa_arrivals = d1[d1['Current (Arrival) District'].str.lower() == 'hargeysa'].groupby('Month End').sum().sort_values(by='Month End').copy()
```

```python
hargesa_arrivals.shape
```

```python
hargesa_arrivals.head()
```

```python
cols = list(hargesa_data.columns)
cols.append('Displacements')
```

```python
hargesa_data = pd.DataFrame(np.hstack((hargesa_data, hargesa_arrivals[' Number of Individuals '].values.reshape(-1,1))), columns=cols)
```

```python
hargesa_data.columns
```

```python
hargesa_data.head(2)
```

```python
final_cols = []
for col in hargesa_data.columns:
    print("*****************************")
    print("Column : %s | Number of missing points: %0.1f | Percent of missing point: %0.1f"% (col, pd.isnull(hargesa_data[col]).sum(), np.round((pd.isnull(hargesa_data[col]).sum()/hargesa_data.shape[0])*100,2)))
    if np.round((pd.isnull(hargesa_data[col]).sum()/hargesa_data.shape[0])*100,2) < 20:
        final_cols.append(col)
```

```python
final_cols
```

```python
final_cols
```

```python
hf = hargesa_data[final_cols].copy()
```

```python
sns.heatmap(hf.isnull())
```

```python
for col in hf.columns:
    if(hf[col].isnull().sum()!=0):
        print(col)
        hf[col].fillna(value = hf[col].mean())
```
